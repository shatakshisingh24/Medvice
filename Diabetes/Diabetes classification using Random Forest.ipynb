```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#Loading the Dataset
data = pd.read_csv('data.csv')
```

```python
#Checking for null values
data.isnull().values.any()
```

```python
data.describe()
```

```python
data.head()
```

```python
#Applying one hot encoder for gender and Diabetes Column
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder() 
  
data['Gender']= le.fit_transform(data['Gender']) 
data['Diabetes']= le.fit_transform(data['Diabetes']) 
data
```

```python
#Separating the features and target variable
X=data.iloc[:,1:15]
Y=data['Diabetes']
```

```python
#Splitting the data into training and test data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.20, random_state=10)
```

```python
#Importing Rnadom forest Classifier for classification
from sklearn.ensemble import RandomForestClassifier
random_forest_model = RandomForestClassifier(random_state=10)
random_forest_model.fit(X_train, y_train.ravel())
```

```python
predictions=random_forest_model.predict(X_test)
from sklearn import metrics
print(metrics.accuracy_score(y_test,predictions))
```

```python
#Printing the accuracy of our Model
print("Accuracy = {0:.3f}".format(metrics.accuracy_score(y_test, predictions)))
```
